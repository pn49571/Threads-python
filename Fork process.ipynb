# Call fork routine and spawn a new program 

```python
import os, time
def counter(count):
    for i in range(count):
        time.sleep(1)
        print('[%s] => %s' % (os.getpid(),i))

for i in range(10):
    pid=os.fork()
    if pid != 0:
        print('Process %d spawned' %pid)
    else:
        counter(10)
        os._exit(0)
print('Main process exciting')
```
